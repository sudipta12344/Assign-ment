```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)  # reproducibility

# Activations + derivatives
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

def dsigmoid(a):
    # input is the activation a = sigmoid(z)
    return a * (1.0 - a)

def relu(z):
    return np.maximum(0.0, z)

def drelu(z):
    return (z > 0).astype(float)

# Metrics from scratch
def confusion_matrix_binary(y_true, y_pred):
    # y_true, y_pred are 0/1 arrays
    tp = np.sum((y_true == 1) & (y_pred == 1))
    tn = np.sum((y_true == 0) & (y_pred == 0))
    fp = np.sum((y_true == 0) & (y_pred == 1))
    fn = np.sum((y_true == 1) & (y_pred == 0))
    return tp, fp, tn, fn

def accuracy_score(y_true, y_pred):
    return np.mean(y_true == y_pred)

def precision_score(y_true, y_pred):
    tp, fp, tn, fn = confusion_matrix_binary(y_true, y_pred)
    return tp / (tp + fp) if (tp + fp) > 0 else 0.0

def recall_score(y_true, y_pred):
    tp, fp, tn, fn = confusion_matrix_binary(y_true, y_pred)
    return tp / (tp + fn) if (tp + fn) > 0 else 0.0

def f1_score(y_true, y_pred):
    p = precision_score(y_true, y_pred)
    r = recall_score(y_true, y_pred)
    return 2*p*r / (p + r) if (p + r) > 0 else 0.0

def roc_curve_from_scores(y_true, y_score, num_thresholds=200):
    # thresholds sweep from 1 to 0 to draw ROC
    thresholds = np.linspace(1.0, 0.0, num_thresholds)
    tpr_list, fpr_list = [], []
    for t in thresholds:
        y_pred = (y_score >= t).astype(int)
        tp, fp, tn, fn = confusion_matrix_binary(y_true, y_pred)
        tpr = tp / (tp + fn) if (tp + fn) > 0 else 0.0
        fpr = fp / (fp + tn) if (fp + tn) > 0 else 0.0
        tpr_list.append(tpr)
        fpr_list.append(fpr)
    return np.array(fpr_list), np.array(tpr_list), thresholds

def auc_trapezoid(x, y):
    # x: FPR ascending, y: TPR; assumes proper ordering
    # Sort by x just in case
    order = np.argsort(x)
    x, y = x[order], y[order]
    return np.trapz(y, x)
```

```python
# XOR dataset that requires a non-linear decision boundary
def make_xor(n=800, noise=0.2):
    X = np.random.uniform(-1.0, 1.0, size=(n, 2))
    y = ((X[:,0] > 0) ^ (X[:,1] > 0)).astype(int)
    # add small Gaussian noise
    X += np.random.normal(0, noise, size=X.shape)
    return X, y

X, y = make_xor(n=800, noise=0.15)

# Train/Val/Test split: 60/20/20
idx = np.arange(len(X))
np.random.shuffle(idx)
X, y = X[idx], y[idx]

n = len(X)
n_train = int(0.6*n)
n_val   = int(0.2*n)

X_train, y_train = X[:n_train], y[:n_train]
X_val,   y_val   = X[n_train:n_train+n_val], y[n_train:n_train+n_val]
X_test,  y_test  = X[n_train+n_val:], y[n_train+n_val:]

X_train.shape, X_val.shape, X_test.shape
```

```python
class MLP221:
    """
    2 inputs -> 2 hidden (ReLU or Sigmoid) -> 1 sigmoid output
    Loss: Mean Squared Error (MSE)
    Optimization: Batch Gradient Descent
    """
    def __init__(self, hidden_activation='relu', lr=0.1):
        assert hidden_activation in ['relu', 'sigmoid']
        self.hidden_activation = hidden_activation
        self.lr = lr
        self._init_params()

    def _init_params(self):
        # Xavier/He-ish small random init
        self.W1 = np.random.randn(2, 2) * 0.5
        self.b1 = np.zeros((1, 2))
        self.W2 = np.random.randn(2, 1) * 0.5
        self.b2 = np.zeros((1, 1))

    def _act_h(self, z):
        if self.hidden_activation == 'relu':
            return relu(z)
        else:
            return sigmoid(z)

    def _dact_h(self, z, a):
        # for sigmoid we can use a*(1-a); for relu we need z>0
        if self.hidden_activation == 'relu':
            return drelu(z)
        else:
            return dsigmoid(a)

    def forward(self, X):
        # X: (N,2)
        self.z1 = X @ self.W1 + self.b1     # (N,2)
        self.a1 = self._act_h(self.z1)      # (N,2)
        self.z2 = self.a1 @ self.W2 + self.b2  # (N,1)
        self.a2 = sigmoid(self.z2)          # output probs (N,1)
        return self.a2

    def loss(self, y_true, y_pred):
        # MSE (mean of 0.5*(y - y_hat)^2)
        y_true = y_true.reshape(-1, 1)
        return np.mean(0.5 * (y_true - y_pred) ** 2)

    def backward(self, X, y_true):
        """
        Backprop for MSE with sigmoid output:
        dL/dz2 = (a2 - y) * sigmoid'(z2) = (a2 - y) * a2*(1-a2)
        """
        N = X.shape[0]
        y_true = y_true.reshape(-1, 1)

        # Output gradient
        da2 = (self.a2 - y_true)                           # d(0.5*(y - a2)^2)/da2 = (a2 - y)
        dz2 = da2 * dsigmoid(self.a2)                      # chain through sigmoid
        dW2 = (self.a1.T @ dz2) / N
        db2 = np.mean(dz2, axis=0, keepdims=True)

        # Hidden layer gradients
        da1 = dz2 @ self.W2.T
        dz1 = da1 * self._dact_h(self.z1, self.a1)
        dW1 = (X.T @ dz1) / N
        db1 = np.mean(dz1, axis=0, keepdims=True)

        # Gradient descent update
        self.W2 -= self.lr * dW2
        self.b2 -= self.lr * db2
        self.W1 -= self.lr * dW1
        self.b1 -= self.lr * db1

    def fit(self, X, y, iters=1000, X_val=None, y_val=None, verbose=False):
        history = {'loss': [], 'val_loss': []}
        for i in range(1, iters+1):
            y_pred = self.forward(X)
            L = self.loss(y, y_pred)
            self.backward(X, y)
            history['loss'].append(L)

            if X_val is not None:
                val_pred = self.forward(X_val)
                history['val_loss'].append(self.loss(y_val, val_pred))

            if verbose and (i % max(1, iters//10) == 0):
                msg = f"[{i}/{iters}] loss={L:.4f}"
                if X_val is not None:
                    msg += f", val_loss={history['val_loss'][-1]:.4f}"
                print(msg)
        return history

    def predict_proba(self, X):
        return self.forward(X).reshape(-1)  # probabilities

    def predict(self, X, threshold=0.5):
        return (self.predict_proba(X) >= threshold).astype(int)
```

```python
learning_rates = [0.01, 0.1, 0.5]
iterations_list = [500, 1000, 5000]
hidden_acts = ['relu', 'sigmoid']

best = None  # to store best config by validation accuracy

for act in hidden_acts:
    for lr in learning_rates:
        for iters in iterations_list:
            model = MLP221(hidden_activation=act, lr=lr)
            model.fit(X_train, y_train, iters=iters, X_val=X_val, y_val=y_val, verbose=False)
            val_pred = model.predict(X_val, threshold=0.5)
            val_acc = accuracy_score(y_val, val_pred)
            cfg = {'act': act, 'lr': lr, 'iters': iters, 'val_acc': val_acc}
            if (best is None) or (val_acc > best['val_acc']):
                best = cfg

print("Best hyperparameters found:")
print(best)
```

```python
# Merge Train + Val for final training
X_trfinal = np.vstack([X_train, X_val])
y_trfinal = np.concatenate([y_train, y_val])

best_model = MLP221(hidden_activation=best['act'], lr=best['lr'])
history = best_model.fit(X_trfinal, y_trfinal, iters=best['iters'], verbose=True)

# Test evaluation
y_scores = best_model.predict_proba(X_test)            # probabilities
y_pred05 = (y_scores >= 0.5).astype(int)               # hard labels at 0.5

acc = accuracy_score(y_test, y_pred05)
prec = precision_score(y_test, y_pred05)
rec = recall_score(y_test, y_pred05)
f1 = f1_score(y_test, y_pred05)

print("\n=== Test Metrics (threshold=0.5) ===")
print(f"Accuracy : {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall   : {rec:.4f}")
print(f"F1-score : {f1:.4f}")

tp, fp, tn, fn = confusion_matrix_binary(y_test, y_pred05)
print(f"Confusion Matrix -> TP:{tp}  FP:{fp}  TN:{tn}  FN:{fn}")
```

```python
fpr, tpr, thresholds = roc_curve_from_scores(y_test, y_scores, num_thresholds=400)
auc = auc_trapezoid(fpr, tpr)

plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, label=f'ROC (AUC={auc:.3f})')
plt.plot([0,1], [0,1], linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve (Test Set)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
def precision_recall_vs_threshold(y_true, y_score, num_thresholds=50):
    ths = np.linspace(0.0, 1.0, num_thresholds)
    precisions, recalls = [], []
    for t in ths:
        y_pred = (y_score >= t).astype(int)
        precisions.append(precision_score(y_true, y_pred))
        recalls.append(recall_score(y_true, y_pred))
    return ths, np.array(precisions), np.array(recalls)

ths, precisions, recalls = precision_recall_vs_threshold(y_test, y_scores, 60)

plt.figure(figsize=(6,4))
plt.plot(ths, precisions, label='Precision')
plt.plot(ths, recalls, label='Recall')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.title('Precision/Recall vs Threshold (Test Set)')
plt.legend()
plt.show()
```
